```python
class Noukataiken():
    
    
    def __init__(self,members=["A","B","C","D"],
                init_budget=500,
                init_mental=200,
                data_path="./datas",):
        self.players = []
        
        self.crops = Crops(data_path)
        self.cards = Cards(data_path)
        self.events = Events(data_path)
        
        self.phase = 0
        self.season = 0
        self.timeseries = 0 
        
        self.history = []
        
        for i in members:
            player_init = Player(i,init_budget,init_mental)
            self.players.append(player_init)
            
    def update(self):
        
        #phase0: title sequence
        if self.phase == 0:

            input("Phase 0: enter the key")
            print("Next")
            self.phase += 1
            return 1
            
        #phase1~ : main sequence
        
        #phase1: planting
        if self.phase == 1:
            for player in self.players:
                crop_name = input("Phase 0 select crop; name:{}".format(player.name))
                player.select_crops(self.crops, crop_name)
            print("Next")
            self.logger()
            print(self.history)
            self.phase += 1
            return 1
            
        #phase2: action
        if self.phase == 2:
            input("Now Phase 2")
            print("Next")
            self.phase += 1
            return 1
        
        #phase3: god
        if self.phase == 3:
            input("Now Phase 3")
            print("Next")
            self.phase += 1
            return 0
        
        #phase4: god
        
    def logger(self):
        for a in self.players:
            status = a.get_status()
            self.history.append(status)
        history_df = pd.DataFrame(self.history)
        print(history_df)
            
            
```

```python
class Player():
    def __init__(self, name, budget, mental):
        self.name   = name
        self.budget = budget
        self.mental = mental
        
        
        self.crop_name = "None"
        self.products = 0
        self.price = 0
        
    def get_status(self):
        return {"プレイヤー名":self.name, "資金":self.budget, "メンタル":self.mental, 
                "作物":self.crop_name, "収量":self.products, "販売価格":self.price}

    def select_crops(self, crops, crop_name):
        crop_data = crops.get_basedata(crop_name)[0]
        print(crop_data)
        if crop_data[0] == 0:
            return -1
        
        if self.mental <= crop_data[5]:
            return -1
        
        self.crop_name = crop_data[1]
        self.products  = crop_data[2]
        self.prices    = crop_data[3]
        
        self.budget   -= crop_data[4]
        
        return 0
        
        
        
```

```python
class Crops():
    def __init__(self, root):
        self.crop_df = pd.read_csv(root + "/Crops.csv")
        
    def get_basedata(self, crop_name):
        if not crop_name in self.crop_df["名称"].values:
            crop_data = [0, "None", 0, 0, 0, 0]
            return crop_data
            
        crop_data = self.crop_df[self.crop_df["名称"] == crop_name].values
        
        return crop_data
```

```python
# TODOアプリ
#--------------------------

import tkinter as tk

# Todoアプリのクラス
class Noukataiken(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Noukataiken")
        self.geometry("400x400")

        # ラベルとエントリーを配置するフレーム
        input_frame = tk.Frame(self)
        input_frame.pack(pady=10)
        tk.Label(input_frame, text="Task:").grid(row=1, column=0)
        self.task_entry = tk.Entry(input_frame, width=30)
        self.task_entry.grid(row=0, column=1)

        # 追加ボタンを配置するフレーム
        btn_frame = tk.Frame(self)
        btn_frame.pack(pady=10)
        tk.Button(btn_frame, text="Add Task", command=self.add_task).pack(side="left")
        tk.Button(btn_frame, text="Delete Task", command=self.delete_task).pack(side="left")

        # タスクの一覧を表示するリストボックス
        self.task_list = tk.Listbox(self, selectmode="multiple", bd=10)
        self.task_list.pack(fill="both", expand=True)

        # タスクの一覧
        self.tasks = []

    # タスクの追加
    def add_task(self):
        task = self.task_entry.get()
        if task:
            self.tasks.append(task)
            self.task_list.insert("end", task)
            self.task_entry.delete(0, "end")

    # タスクの削除
    def delete_task(self):
        selected_indices = self.task_list.curselection()
        for index in selected_indices[::-1]:
            self.task_list.delete(index)
            self.tasks.pop(index)

# Todoアプリを起動する
if __name__ == "__main__":
    app = TodoApp()
    app.mainloop()
```
